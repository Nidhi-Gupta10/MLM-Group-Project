## Contingency Table

### Expected Frequencies:
| Shipping Method | 0.0       | 1.0       | 2.0       |
|------------------|-----------|-----------|-----------|

## Chi-Square Test Results:
- **Chi-squared Statistic**: 7.49
- **P-value**: 0.024
- **Degrees of Freedom (df)**: 2

## Interpretation:
1. **Hypotheses**:
   - **Null Hypothesis (H₀)**: There is no association between "Shipping Method" and "Import_Export".
   - **Alternative Hypothesis (H₁)**: There is an association between "Shipping Method" and "Import_Export".

2. **Significance Level (α)**: 0.05
   - Since the p-value (0.024) is less than the significance level (0.05), we **reject the null hypothesis**.

3. **FINDINGS**:
   - There is a **statistically significant association** between "Shipping Method" and "Import_Export". This suggests that the choice of shipping method depends on whether the shipment is an import or export.

```python
# Handling Missing Data
def missing_data_summary(df):
    """
    Creates a DataFrame summarizing missing data statistics.

    Args:
        df: The input DataFrame.

    Returns:
        A DataFrame with missing data count and percentage for each variable and row.
    """
    # Column-wise summary
    missing_count_col = df.isnull().sum()
    missing_percentage_col = (missing_count_col / len(df)) * 100
    missing_summary_col = pd.DataFrame({
        'Missing Count': missing_count_col,
        'Missing Percentage': missing_percentage_col
    })
    # Row-wise summary
    missing_count_row = df.isnull().sum(axis=1)
    missing_percentage_row = (missing_count_row / len(df.columns)) * 100
    missing_summary_row = pd.DataFrame({
        'Missing Count': missing_count_row,
        'Missing Percentage': missing_percentage_row
    }, index=df.index)  # Use original DataFrame index

    return missing_summary_col, missing_summary_row

# Calculate missing data summary for categorical subset
ng06_missing_summary_cat_col, ng06_missing_summary_cat_row = missing_data_summary(ng06_ds_cat)

# Calculate missing data summary for non-categorical subset
ng06_missing_summary_noncat_col, ng06_missing_summary_noncat_row = missing_data_summary(ng06_ds_noncat)

# Drop columns and rows with more than 50% missing data for categorical subset
ng06_ds_cat_treated = ng06_ds_cat.dropna(thresh=len(ng06_ds_cat) * 0.5, axis=1)
ng06_ds_cat_treated = ng06_ds_cat_treated.dropna(thresh=len(ng06_ds_cat_treated.columns) * 0.5, axis=0)

# Drop columns and rows with more than 50% missing data for non-categorical subset
ng06_ds_noncat_treated = ng06_ds_noncat.dropna(thresh=len(ng06_ds_noncat) * 0.5, axis=1)
ng06_ds_noncat_treated = ng06_ds_noncat_treated.dropna(thresh=len(ng06_ds_noncat_treated.columns) * 0.5, axis=0)

# Calculate total variables and rows dropped
ng06_dropped_variables_cat = len(ng06_ds_cat.columns) - len(ng06_ds_cat_treated.columns)
ng06_dropped_rows_cat = len(ng06_ds_cat) - len(ng06_ds_cat_treated)
ng06_dropped_variables_noncat = len(ng06_ds_noncat.columns) - len(ng06_ds_noncat_treated.columns)
ng06_dropped_rows_noncat = len(ng06_ds_noncat) - len(ng06_ds_noncat_treated)

ng06_total_dropped_variables = ng06_dropped_variables_cat + ng06_dropped_variables_noncat
ng06_total_dropped_rows = ng06_dropped_rows_cat + ng06_dropped_rows_noncat

# Print missing data summaries
print("Categorical Missing Data Summary (Columns):")
print(ng06_missing_summary_cat_col)
print("\nCategorical Missing Data Summary (Rows):")
print(ng06_missing_summary_cat_row)
print("\nNon-Categorical Missing Data Summary (Columns):")
print(ng06_missing_summary_noncat_col)
print("\nNon-Categorical Missing Data Summary (Rows):")
print(ng06_missing_summary_noncat_row)

# Print total dropped variables and rows
print(f"\nTotal variables dropped: {ng06_total_dropped_variables}")
print(f"Total rows dropped: {ng06_total_dropped_rows}")
```

## Descriptive Statistics

```python
# Function to calculate descriptive statistics for categorical variables
def ng06_categorical_stats(df, cat_columns):
    ng06_results = []

    for ng06_col in cat_columns:
        ng06_counts = df[ng06_col].value_counts()
        ng06_proportions = ng06_counts / len(df)
        ng06_mode = ng06_counts.idxmax()
        ng06_min = df[ng06_col].min()
        ng06_max = df[ng06_col].max()

        for category, count in ng06_counts.items():
            ng06_results.append({
                "Column": ng06_col,
                "Category": category,
                "Frequency": count,
                "Proportion": ng06_proportions[category],
                "Minimum": ng06_min,
                "Maximum": ng06_max,
                "Mode": ng06_mode
            })

    return pd.DataFrame(ng06_results)

# Function to calculate correlation for categorical columns
def ng06_categorical_correlation(df, cat_columns, method='spearman'):
    ng06_corr_results = pd.DataFrame(index=cat_columns, columns=cat_columns, dtype=float)

    for i in range(len(cat_columns)):
        for j in range(i, len(cat_columns)):
            ng06_col1, ng06_col2 = cat_columns[i], cat_columns[j]
            if i == j:
                # Perfect correlation for the same variable
                corr = 1.0
            else:
                if method == 'spearman':
                    corr, _ = spearmanr(df[ng06_col1].astype('category').cat.codes,
                                        df[ng06_col2].astype('category').cat.codes)
                elif method == 'kendall':
                    corr, _ = kendalltau(df[ng06_col1].astype('category').cat.codes,
                                         df[ng06_col2].astype('category').cat.codes)
            ng06_corr_results.loc[ng06_col1, ng06_col2] = corr
            ng06_corr_results.loc[ng06_col2, ng06_col1] = corr

    return ng06_corr_results

# Retrieve categorical columns from ng06_ds_cat_encoded
ng06_cat_columns = ng06_ds_cat_encoded.columns.tolist()

# Generate descriptive stats and correlations for these columns
ng06_categorical_stats_df = ng06_categorical_stats(ng06_ds_preprocessed, ng06_cat_columns)
ng06_spearman_corr_df = ng06_categorical_correlation(ng06_ds_preprocessed, ng06_cat_columns, method='spearman')
ng06_kendall_corr_df = ng06_categorical_correlation(ng06_ds_preprocessed, ng06_cat_columns, method='kendall')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
ng06_spearman_corr_df.fillna(1, inplace=True)
ng06_kendall_corr_df.fillna(1, inplace=True)

# Display results
print("Descriptive Statistics:")
print(ng06_categorical_stats_df)  # Show all rows

print("\nSpearman Correlation:")
print(ng06_spearman_corr_df)

print("\nKendall Correlation:")
print(ng06_kendall_corr_df)
```

```python
# Descriptive Stats for Noncat Variables

# Function to calculate descriptive statistics for numerical variables
def ng06_numerical_stats(df, num_columns):
    ng06_results = []

    for ng06_col in num_columns:
        ng06_data = df[ng06_col].dropna()  # Exclude NaN values
        ng06_min = ng06_data.min()
        ng06_max = ng06_data.max()
        ng06_mean = ng06_data.mean()
        ng06_median = ng06_data.median()
        ng06_mode = ng06_data.mode().iloc[0] if not ng06_data.mode().empty else np.nan
        ng06_range = ng06_max - ng06_min
        ng06_std_dev = ng06_data.std()
        ng06_skewness = skew(ng06_data)
        ng06_kurt = kurtosis(ng06_data)

        # Coefficient of Variation (CV)
        ng06_cv = ng06_std_dev / ng06_mean if ng06_mean != 0 else np.nan

        # Confidence Interval (95%)
        ng06_conf_interval = norm.interval(0.95, loc=ng06_mean, scale=ng06_std_dev / np.sqrt(len(ng06_data)))

        # Add results
        ng06_results.append({
            "Column": ng06_col,
            "Minimum": ng06_min,
            "Maximum": ng06_max,
            "Mean": ng06_mean,
            "Median": ng06_median,
            "Mode": ng06_mode,
            "Range": ng06_range,
            "Standard Deviation": ng06_std_dev,
            "Skewness": ng06_skewness,
            "Kurtosis": ng06_kurt,
            "Coefficient of Variation": ng06_cv,
            "Confidence Interval (95%)": ng06_conf_interval
        })

    return pd.DataFrame(ng06_results)

# Function to calculate correlation for numerical variables
def ng06_numerical_correlation(df, num_columns, method='pearson'):
    ng06_corr_results = pd.DataFrame(index=num_columns, columns=num_columns, dtype=float)

    for i in range(len(num_columns)):
        for j in range(i, len(num_columns)):
            ng06_col1, ng06_col2 = num_columns[i], num_columns[j]
            if i == j:
                corr = 1.0  # Perfect correlation for the same variable
            else:
                if method == 'pearson':
                    corr, _ = pearsonr(df[ng06_col1].dropna(), df[ng06_col2].dropna())
                elif method == 'spearman':
                    corr, _ = spearmanr(df[ng06_col1].dropna(), df[ng06_col2].dropna())
            ng06_corr_results.loc[ng06_col1, ng06_col2] = corr
            ng06_corr_results.loc[ng06_col2, ng06_col1] = corr

    return ng06_corr_results

# Retrieve numerical columns from ng06_ds_noncat_scaled
ng06_num_columns = ng06_ds_noncat_scaled.columns.tolist()

# Generate statistics and correlations for these columns
ng06_numerical_stats_df = ng06_numerical_stats(ng06_ds_noncat_scaled, ng06_num_columns)
ng06_pearson_corr_df = ng06_numerical_correlation(ng06_ds_noncat_scaled, ng06_num_columns, method='pearson')
ng06_spearman_corr_df = ng06_numerical_correlation(ng06_ds_noncat_scaled, ng06_num_columns, method='spearman')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
ng06_pearson_corr_df.fillna(1, inplace=True)
ng06_spearman_corr_df.fillna(1, inplace=True)

# Display results
print("Numerical Statistics:")
print(ng06_numerical_stats_df)

print("\nPearson Correlation:")
print(ng06_pearson_corr_df)

print("\nSpearman Correlation:")
print(ng06_spearman_corr_df)
```

```python
# Assuming your preprocessed dataset is named 'ng06_ds_preprocessed'
# and your non-categorical columns are 'Quantity', 'Value', 'Weight'

# List of non-categorical variables
ng06_non_categorical_vars = ['Quantity', 'Value', 'Weight']

# Loop through each variable
for ng06_var in ng06_non_categorical_vars:
    print(f"\n--- Tests for {ng06_var} ---")

    # Shapiro-Wilk Test
    ng06_shapiro_stat, ng06_shapiro_p_value = stats.shapiro(ng06_ds_preprocessed[ng06_var])
    print(f"Shapiro-Wilk Test - Statistic: {ng06_shapiro_stat}, P-value: {ng06_shapiro_p_value}")
    if ng06_shapiro_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Kolmogorov-Smirnov Test
    ng06_ks_stat, ng06_ks_p_value = stats.kstest(ng06_ds_preprocessed[ng06_var], 'norm')
    print(f"Kolmogorov-Smirnov Test - Statistic: {ng06_ks_stat}, P-value: {ng06_ks_p_value}")
    if ng06_ks_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Anderson-Darling Test
    ng06_anderson_result = stats.anderson(ng06_ds_preprocessed[ng06_var])
    print("Anderson-Darling Test - Statistic:", ng06_anderson_result.statistic)
    # Check only for 5% significance level
    if ng06_anderson_result.statistic < ng06_anderson_result.critical_values[2]:  # Index 2 corresponds to 5%
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
    else:
        print(f"Data does not look normal (reject H0) at 5% significance level")

    # Jarque-Bera Test
    ng06_jb_stat, ng06_jb_p_value = stats.jarque_bera(ng06_ds_preprocessed[ng06_var])
    print(f"Jarque-Bera Test - Statistic: {ng06_jb_stat}, P-value: {ng06_jb_p_value}")
    if ng06_jb_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
```

### Clustering

```python
# Select relevant columns for clustering
ng06_cluster_data = ng06_ds_preprocessed[['Value','Import_Export','Quantity']]

# Store cluster models in a dictionary
ng06_cluster_models = {}

# Loop through different cluster numbers
for ng06_n_clusters in range(2, 6):
    # K-Means Clustering
    ng06_kmeans = KMeans(n_clusters=ng06_n_clusters, random_state=11006)
    ng06_kmeans_labels = ng06_kmeans.fit_predict(ng06_cluster_data)
    ng06_cluster_models[f"ng06_kmeans_{ng06_n_clusters}cluster"] = ng06_kmeans

    # Agglomerative Hierarchical Clustering
    ng06_agg = AgglomerativeClustering(n_clusters=ng06_n_clusters, metric='euclidean', linkage='ward')
    ng06_agg_labels = ng06_agg.fit_predict(ng06_cluster_data)
    ng06_cluster_models[f"ng06_agg_{ng06_n_clusters}cluster"] = ng06_agg

    # BIRCH Clustering
    ng06_birch = Birch(n_clusters=ng06_n_clusters)
    ng06_birch_labels = ng06_birch.fit_predict(ng06_cluster_data)
    ng06_cluster_models[f"ng06_birch_{ng06_n_clusters}cluster"] = ng06_birch

    # Create a DataFrame with values and cluster labels for each model
    for model_name, labels in [('KMeans', ng06_kmeans_labels),
                                ('Agglomerative', ng06_agg_labels),
                                ('BIRCH', ng06_birch_labels)]:
        ng06_cluster_df = pd.DataFrame(ng06_cluster_data)
        ng06_cluster_df['Cluster'] = labels

        # Display the DataFrame
        print(f"{model_name} Clusters for {ng06_n_clusters} clusters:")
        print(ng06_cluster_df.head(), '\n')
```

```python
# Assuming your results DataFrame is named 'ng06_results_df'

# Function to find the best cluster number for a given model
def find_best_cluster_number(df, model_name, metric='Silhouette Score'):
    """
    Finds the best cluster number for a given model based on a specified metric.

    Args:
        df: The DataFrame containing the clustering results.
        model_name: The name of the clustering model.
        metric: The metric to use for evaluation ('Silhouette Score' or 'DB Score').

    Returns:
        The best cluster number.
    """
    # Filter results for the specified model
    model_results = df[df['Model'] == model_name]

    # Find the cluster number with the best score
    if metric == 'Silhouette Score':
        best_cluster_number = model_results.loc[model_results['Silhouette Score'].idxmax(), 'Number of Clusters']
    elif metric == 'DB Score':
        best_cluster_number = model_results.loc[model_results['DB Score'].idxmin(), 'Number of Clusters']
    else:
        raise ValueError("Invalid metric. Choose 'Silhouette Score' or 'DB Score'.")

    return best_cluster_number

# Find the best cluster number for each model
best_kmeans_cluster_silhouette = find_best_cluster_number(ng06_results_df, 'ng06_kmeans', metric='Silhouette Score')
best_kmeans_cluster_db = find_best_cluster_number(ng06_results_df, 'ng06_kmeans', metric='DB Score')
best_agg_cluster_silhouette = find_best_cluster_number(ng06_results_df, 'ng06_agg', metric='Silhouette Score')
best_agg_cluster_db = find_best_cluster_number(ng06_results_df, 'ng06_agg', metric='DB Score')
best_birch_cluster_silhouette = find_best_cluster_number(ng06_results_df, 'ng06_birch', metric='Silhouette Score')
best_birch_cluster_db = find_best_cluster_number(ng06_results_df, 'ng06_birch', metric='DB Score')

# Print the results
print(f"Best cluster number for KMeans (Silhouette): {best_kmeans_cluster_silhouette}")
print(f"Best cluster number for KMeans (DB Score): {best_kmeans_cluster_db}")
print(f"Best cluster number for Agglomerative (Silhouette): {best_agg_cluster_silhouette}")
print(f"Best cluster number for Agglomerative (DB Score): {best_agg_cluster_db}")
print(f"Best cluster number for BIRCH (Silhouette): {best_birch_cluster_silhouette}")
print(f"Best cluster number for BIRCH (DB Score): {best_birch_cluster_db}")
```

```python
# Selecting the DBSCAN model (assuming it's stored in ng06_dbscan)
print("Selecting DBSCAN model")

# Use the pre-calculated DBSCAN labels (assuming they are stored in ng06_dbscan_labels)
ng06_selected_cluster_labels = ng06_dbscan_labels

# Create a DataFrame with cluster labels for the selected model
ng06_selected_cluster_df = pd.DataFrame({'Cluster': ng06_selected_cluster_labels}, index=ng06_ds_preprocessed.index)

# Merge cluster labels with preprocessed data
ng06_data_for_anova = ng06_ds_preprocessed.merge(ng06_selected_cluster_df[['Cluster']], left_index=True, right_index=True, how='left')

# Automatically identify columns used for clustering (using your ng06_ds_noncat_scaled columns)
ng06_cluster_columns = ng06_ds_noncat_scaled.columns

# Import pairwise_tukeyhsd for post-hoc test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Loop through the cluster columns
for ng06_var in ng06_cluster_columns:
    # Create groups for ANOVA
    ng06_groups = [ng06_data_for_anova[ng06_var][ng06_data_for_anova['Cluster'] == i] for i in ng06_data_for_anova['Cluster'].unique()]

    # Perform ANOVA using f_oneway
    ng06_fvalue, ng06_pvalue = stats.f_oneway(*ng06_groups)

    # Print ANOVA results
    print(f"ANOVA for {ng06_var}:")
    print(f"F-statistic: {ng06_fvalue:.2f}")
    print(f"P-value: {ng06_pvalue:.3f}")
    print("-" * 50)

    # Perform Tukey's HSD post-hoc test if ANOVA is significant
    if ng06_pvalue < 0.05:  # Check for significance
        # Perform Tukey's HSD post-hoc test
        ng06_tukey_result = pairwise_tukeyhsd(ng06_data_for_anova[ng06_var], ng06_data_for_anova['Cluster'], alpha=0.05)

        print(f"Tukey's HSD Post-Hoc Test for {ng06_var}:")
        print(ng06_tukey_result)
        print("-" * 50)
```

```python
# Frequency Analysis for Categorical Data with DBSCAN:


# Create a DataFrame with cluster labels for DBSCAN
ng06_dbscan_cluster_df = pd.DataFrame({'Cluster': ng06_dbscan_labels}, index=ng06_ds_preprocessed.index)

# Merge cluster assignments with preprocessed data
ng06_data_with_clusters = ng06_ds_preprocessed.merge(ng06_dbscan_cluster_df[['Cluster']], left_index=True, right_index=True, how='left')

# List of categorical variables (adjust as needed)
freq_columns = ['Import_Export', 'Category', 'Shipping_Method', 'Payment_Terms']

# Loop through each categorical variable
for var in freq_columns:
    # Create a cross-tabulation table
    freq_table = pd.crosstab(ng06_data_with_clusters['Cluster'], ng06_data_with_clusters[var])

    # Print the frequency table
    print(f"Frequency Table for {var} (DBSCAN):")
    print(freq_table)
    print("-" * 50)

    # Perform Chi-squared test for independence
    chi2, p, dof, expected = stats.chi2_contingency(freq_table)
    print(f"Chi-squared test for {var} (DBSCAN):")
    print(f"Chi-squared statistic: {chi2}")
    print(f"P-value: {p}")
    print("-" * 50)
```
